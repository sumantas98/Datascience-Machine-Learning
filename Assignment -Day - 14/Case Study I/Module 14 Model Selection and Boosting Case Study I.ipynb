```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split,KFold,cross_val_score
from sklearn.ensemble import AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
data = pd.read_csv('bio-degradabale-data.csv',sep=';',header=None)
```

```python
data.head()
```

```python
data.dropna()
```

```python
x=data.iloc[:,0:41]
y=data.iloc[:,41:]
```

```python
dict_model = DecisionTreeClassifier()
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=.20,random_state=1010)
```

```python
dict_model.fit(train_x,train_y)
```

```python
y_predict=dict_model.predict(test_x)
print('Accuary Score',metrics.accuracy_score(y_predict,test_y)*100)
```

```python
kfold = KFold(n_splits=50,random_state=1010)
adf_model = AdaBoostClassifier(DecisionTreeClassifier(max_depth=5)
                               ,n_estimators=45,random_state=1010)
print('accuracy ::: ',cross_val_score(adf_model,x,y,cv=kfold).mean())
```
